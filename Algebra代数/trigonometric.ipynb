```python
# 三角函数
import math
import numpy as np
import matplotlib.pyplot as plt

def deg2rad(deg: int):
  rad = deg*np.pi/180
  return rad

def rad2deg(rad: int):
  deg = rad/np.pi*180
  return deg

x = 100
print(deg2rad(x))

rad = 2
print(rad2deg(rad))
```

```python
def my_sin(x):
  return np.sin(x)

def my_cos(x):
  return np.cos(x)

def my_tan(x):
  return np.tan(x)

nx = np.linspace(-np.pi, np.pi)
nx2 = nx + deg2rad(-45)

plt.style.use('ggplot')

fig, ax = plt.subplots(dpi=80)
ax.plot(nx, my_sin(nx), label='sin')
ax.plot(nx, my_cos(nx), label='cos')
ax.plot(nx, my_cos(nx2), label='offset')
ax.plot(nx, my_tan(nx), label='tan')
ax.legend()
ax.set_xlabel('radian')
ax.set_ylabel('y')
ax.set_ylim(-1.3, 1.3)

```
