```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
arr = np.linspace(-5, 5, 10000)
# print(arr)
x = arr
y = np.sqrt(25-np.square(x))
print(y)
y2 = -y
print(y2)

fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x, y, label='a')
p2, = ax.plot(x, y2, label='b')
p2.set_linestyle('--')
ax.set_xlabel("X Value", size=14)
ax.set_ylabel("y Value", size=10)
ax.legend()
```
