```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
import matplotlib
```

```python
data1 = np.random.randn(100)
# print(data1)
xdata = np.arange(len(data1))
print(xdata)

fig, axs = plt.subplots(2, 1)
plt.subplots_adjust(hspace=0.5)

axs[0].plot(xdata, data1)
axs[0].set_title('Automatic Tick')


axs[1].plot(xdata, data1)
axs[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
axs[1].set_yticks([-1.5, 0, 1.5])
axs[1].set_title('Manual ticks')
axs[1].tick_params(axis='y', size=10, right='')
axs[1].xaxis.set_major_formatter(ticker.FormatStrFormatter(':%.1f:'))
```

```python
fig, ax = plt.subplots(figsize=(10, 5.7))
dates = np.arange(np.datetime64('2022-05-01'), np.datetime64('2022-07-02'), np.timedelta64(1, 'h'))
# print(dates)
data = np.cumsum(np.random.randn(len(dates)))
ax.plot(dates, data)
cdf = matplotlib.dates.ConciseDateFormatter(ax.xaxis.get_major_locator())
print(cdf)
ax.xaxis.set_major_formatter(cdf)
```

```python
fig, ax = plt.subplots(dpi=100)
categories = ['football', 'basketball', 'rugby', 'f1']
bar = ax.bar(categories, np.random.rand(len(categories)), label='a')
print(fig, ax, bar)
```

```python
t = np.arange(0, 5, 0.01)
# print(t)
s = np.cos(2*t*np.pi)
# print(s)

#  
# 1. twin 
fig, (ax1, ax3) = plt.subplots(1, 2, figsize=(13, 5))
l1, = ax1.plot(t, s,)
ax2 = ax1.twinx()
print(ax1, ax2)
l2, = ax2.plot(t, range(len(t)), 'C1') # c1 
ax2.legend([l1, l2], ['Sin(Left)', 'Straight(Right)'])

# 2. secondary 
ax3.plot(t, s)
ax3.set_xlabel('Angle[rad]')
ax4 = ax3.secondary_xaxis('top', functions=(np.rad2deg, np.deg2rad))  # 
ax4.set_xlabel('Angle[]')
```
