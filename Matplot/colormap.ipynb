```python
# 色彩图 Colormap
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mat
```

```python
nx, ny = (3, 2)
print(nx, ny)
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
print(x, y)

xv, yv = np.meshgrid(x, y, indexing='xy')
print(xv)
print(yv)

```

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)


fig, axs = plt.subplots(2, 2, dpi=100)
fig.subplots_adjust(top=1.4, right=1.2)
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')

co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

pc = axs[1, 0].imshow(Z**2 * 100, cmap='plasma',
                      norm=mat.colors.LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1, 0], extend='both')
axs[1, 0].set_title('imshow() with LogNorm()')

data1, data2, data3 = np.random.randn(3, 100)

pc = axs[1, 1].scatter(data1, data2, c=data3, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()')
```

```python
nx, ny = (10, 10)

X = np.linspace(0, 1, nx)
Y = np.linspace(0, 1, ny)
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

xv, yv = np.meshgrid(X, Y)
zv = xv + yv
# print(xv[0])
# print(zv)
fig, axs = plt.subplots(1, 2, figsize=(13, 5.7))
axs[0].scatter(xv, yv)
axs[0].set_xlim(-.1, 1.1)


axs[1].pcolormesh(xv, yv, zv, vmin=0, vmax=2, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1])
```
