# 微分入门

`函数 --> 极限 --> 导数 --> 微分`

所谓微分，简而言之就是**变化的比例**。例如，用时间对移动物体的位置进行微分就可以得到该物体的速度。

通过学习复杂函数的微分，可以预测参数对整体的影响。

### 函数

函数是数学中的一种映射关系，它将一个集合中的元素映射到另一个集合中的元素。例如，函数 $f(x) = x^2$ 将实数集映射到实数集，其中 $x$ 是输入，$f(x)$ 是输出。

函数可以表示为 $y = f(x)$，其中 $y$ 是因变量，$x$ 是自变量。函数的图像可以表示为 $y$ 对 $x$ 的关系。

函数的属性包括：

- 单调性：函数在某区间内单调递增或单调递减。例如，函数 $f(x) = x^2$ 在整个实数集上是单调递增的。
- 奇偶性：函数是奇函数还是偶函数。例如，函数 $f(x) = x^3$ 是奇函数，函数 $f(x) = x^2$ 是偶函数。
- 周期性：函数是否有周期性。例如，函数 $f(x) = \sin(x)$ 是周期函数，其周期为 $2\pi$。

函数的运算包括：

- 加法：$f(x) + g(x)$，例如 $f(x) = x^2 + 2x + 1$。
- 减法：$f(x) - g(x)$，例如 $f(x) = x^2 - 2x + 1$。
- 乘法：$f(x) \cdot g(x)$，例如 $f(x) = x^2 \cdot 2x$。
- 除法：$\frac{f(x)}{g(x)}$，例如 $f(x) = \frac{x^2}{2x}$。


# 极限与微分

## 极限

极限是数学中的一种概念，用于描述函数在某个点附近的值如何变化。具体来说，极限描述的是当自变量趋近于某个值时，函数的值趋近于什么。

例如，对于函数f(x) = x^2，当x趋近于0时，f(x)的值趋近于0。这个极限可以用符号表示为：

lim(x→0) x^2 = 0

这个极限表示当x趋近于0时，f(x)的值趋近于0。
