```python
# K近邻算法 KNeighborClassification
import numpy as np
import pandas as pd
import mglearn
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
```

```python
# 引入数据集
iris_dataset = load_iris()

# 划分数据集
res = train_test_split(iris_dataset.data, iris_dataset.target, random_state=0)
X_train, X_test, y_train, y_test = res

# 格式转换
iris_df = pd.DataFrame(data=X_train, columns=iris_dataset.feature_names)
print(iris_df)

# 绘制散点矩阵
grr = pd.plotting.scatter_matrix(iris_df, c=y_train, figsize=(20, 20), marker='o', diagonal='hist', 
                                 hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
# 引入模型训练

# 实例化k紧邻分类算法
knn = KNeighborsClassifier(n_neighbors=1)

# fit 训练
knn.fit(X_train, y_train)

# predict 根据模型进行新数据预测
X_new = np.array([[5, 3, 3, 2]])
print(f"X_new shape: {X_new.shape}")

prediction = knn.predict(X_new)
print(f"Prediction target: {prediction}; Target name: {iris_dataset.target_names[prediction]}")
```
